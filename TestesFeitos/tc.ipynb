.. code:: python module
    
    from tbx_perftest.tester import Tester
    import numpy as np
    
    def assert_fnc(user_fnc, input, groundtruth):
        f = input
        g = user_fnc(f)
        ok1 = float(abs(g - groundtruth).max() <= 1.)
        ok2 = 1. - float(np.may_share_memory(f,g))
        #return ok1
        return (ok1 + ok2)/2
    tester = Tester(assert_fnc, 'tbx_perftest', 'pirp_4e_tc',runs=25)
