Criando o teste

.. code:: python module
    
    from tbx_perftest.tester import Tester
    
    def assert_fnc(user_fnc, input, groundtruth):
        f = input
        return float(abs(user_fnc(f) - groundtruth).max() <= 1.)
    
    tester = Tester(assert_fnc, 'tbx_perftest', 'pirp_4e_sb',runs=20)

    from tbx_perftest.pirp_4e_sb import tester
    import numpy as np
    import ia636 as ia
    
    # IMPORTANTE: Os casos de teste devem ser apagados se esta página sofrer múltiplas edições
    tester.reset_cases()
    
    def sb_gabarito(f):
        import ia636
        my = np.array([[1, 2, 1],
                       [0, 0, 0],
                       [-1,-2,-1]])
        mx = np.array([[1, 0, -1],
                       [2, 0, -2],
                       [1, 0, -1]])
        gy = ia636.iaconv(f,my)
        gx = ia636.iaconv(f,mx)
        g = np.floor(np.sqrt(gx*gx + gy*gy))
        return g
    
    # -----------------------------------
    # Registra cada procedimento de teste
    #     def addTester(descr, fun, args=[])
    # -----------------------------------
    
    f = np.array([[  0,  0,  0, 200, 200, 200, 0, 0, 0],
                  [  0,  0,  0, 200, 200, 200, 0, 0, 0],
                  [100,100,100, 200, 200, 200,50,50,50],
                  [100,100,100, 200, 200, 200,50,50,50],
                  [100,100,100, 200, 200, 200,50,50,50],
                  [  0,  0,  0, 200, 200, 200, 0, 0, 0],
                  [  0,  0,  0, 200, 200, 200, 0, 0, 0],
                  [  0,  0,  0, 200, 200, 200, 0, 0, 0]])
    f1 = np.array([[   0,  0,  0],
                   [   0,  0,  0],
                   [ 100,100,100]])
    img   = adreadgray('cameraman.tif')
    casos = [\
         [f1, 'num. 1'],
         [f,'numérico'],
         [img, 'imagem'],
         ]
    
    for a,msg in casos:
        tester.add_test_case(a, sb_gabarito(a), label=msg, show_in_results_table=True)
        tester.add_train_case(a, sb_gabarito(a))
