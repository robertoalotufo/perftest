Criando o teste
===============

.. code:: python module
    
    from tbx_perftest.tester import Tester
    
    def assert_fnc(user_fnc, input, groundtruth):
        f,t = input
        return float(abs(user_fnc(f,t) - groundtruth).max() <= 1.)
    
    tester = Tester(assert_fnc, 'tbx_perftest', 'pirp_4e_tp',runs=20)

    from tbx_perftest.pirp_4e_tp import tester
    import numpy as np
    import ia636 as ia
    
    # IMPORTANTE: Os casos de teste devem ser apagados se esta página sofrer múltiplas edições
    tester.reset_cases()
    
    def tp_gabarito(f,t):
        import ia636
        g = ia636.iaptrans(f,t)
        return g
    
    # -----------------------------------
    # Registra cada procedimento de teste
    #     def addTester(descr, fun, args=[])
    # -----------------------------------
    
    f = arange(50).reshape((5,10)).astype(uint8)
    img   = adreadgray('cameraman.tif')
    casos = [\
         [(f,(2,3)),'num. positivo'],
         [(f,(3,-1)),'num. negativo'],
         [(f,(13,-12)), 'num. > que H,W'],
         [(img,(10,100)), 'imagem'],
         ]
    
    for a,msg in casos:
        tester.add_test_case(a, tp_gabarito(*a), label=msg, show_in_results_table=True)
        tester.add_train_case(a, tp_gabarito(*a))
