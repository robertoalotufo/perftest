Criando o teste
===============

.. code:: python module
    
    from tbx_perftest.tester import Tester
    import numpy as np
    
    def assert_fnc(user_fnc, input, groundtruth):
        g = user_fnc(input)
        ok_view = float(g.base is input)
        ok      = float(np.array_equal(g,groundtruth))
        return (ok_view + ok)/2.
    
    tester = Tester(assert_fnc, 'tbx_perftest', 'pirp_4e_cr',runs=10)

.. code:: python
    
    from tbx_perftest.pirp_4e_cr import tester
    import numpy as np
    import ia636
    
    # IMPORTANTE: Os casos de teste devem ser apagados se esta página sofrer múltiplas edições
    tester.reset_cases()
    
    def cr_gabarito(f):
        ci = np.nonzero(f.max(0))[0]
        ri = np.nonzero(f.max(1))[0]
        return f[ri[0]:ri[-1]+1, ci[0]:ci[-1]+1].copy()
    
    # -----------------------------------
    # Registra cada procedimento de teste
    #     def addTester(descr, fun, args=[])
    # -----------------------------------

    f = np.pad(np.ones((5,8),'uint8'),(3,),'constant', constant_values=(0,))    
    f[3,3]=0
    tester.add_test_case(f, cr_gabarito(f), label='numerico', show_in_results_table=True)
    gc1 = ia636.iacircle((201,220), 40, (40,100))
    gc2 = ia636.iacircle((201,220), 10, (80,40))
    g = np.pad(maximum(gc1,gc2),((24,63),(85,38)),'constant', constant_values=(0,))
    tester.add_test_case(g,  cr_gabarito(g) , label='imagem',   show_in_results_table=True)
    
    tester.add_train_case(f, cr_gabarito(f))
    tester.add_train_case(g, cr_gabarito(g))
