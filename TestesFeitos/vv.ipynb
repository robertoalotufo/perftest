Criando o teste
===============

.. code:: python module
    
    import numpy as np
    from tbx_perftest.tester import Tester
    
    def assert_fnc(user_fnc, input, groundtruth):
        X,i,s = input
        return float(abs(user_fnc(X,i,s) - groundtruth).max() <= np.finfo(np.float32).eps)
    
    tester = Tester(assert_fnc, 'tbx_perftest', 'pirp_4e_vv', runs = 10)
