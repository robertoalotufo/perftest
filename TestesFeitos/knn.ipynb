    import numpy as np
    from tbx_perftest.tester import Tester
    
    def assert_fnc(user_fnc, input, groundtruth):
        Xtest,Y,X,k = input
        Ytest = groundtruth
        Yt = user_fnc(Xtest,Y,X,k)
        if (Yt.shape == Ytest.shape):
            return 1. * (Yt == Ytest).sum()/Yt.size
        else:
            return 0.
    
    tester = Tester(assert_fnc, 'tbx_perftest', 'pirp_4e_knn', runs=3)

    def knn_gabarito(Xtest,Y,X,k):
        # input: Xtest, amostras a serem classificadas, dimensões (160,)
        #        Y, Rótulo das classes das amostras X de treinamento (80,)
        #        X, Amostras de treinamento (80,160)
        #        k, n. de vizinhos do classificador k-vizinhos
        # output:y, classe da amostra x, inteiro de 0 a 9
        import numpy as np
        import ia636 as ia
        Ytest = np.zeros((Xtest.shape[0],),'uint8')
        for i,x in enumerate(Xtest):
            x = x.astype(float)
            dd =  ((X-x) * (X-x)).sum(axis=1)
            ddi = np.argsort(dd)[:k]
            h = ia.iahistogram(Y[ddi])
            Ytest[i] = np.argmax(h)
        return Ytest
