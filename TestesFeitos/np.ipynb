Criando o teste
===============

.. code:: python module
    
    from tbx_perftest.tester import Tester
    
    def assert_fnc(user_fnc, input, groundtruth):
        f, perc = input
        return float(abs(user_fnc(f,perc) - groundtruth).max() <= 1.)
    
    tester = Tester(assert_fnc, 'tbx_perftest', 'pirp_4e_np',10)

    from tbx_perftest.pirp_4e_np import tester
    import numpy as np
    import ia636 as ia
    
    # IMPORTANTE: Os casos de teste devem ser apagados se esta página sofrer múltiplas edições
    tester.reset_cases()
    
    def nn_gabarito(f, p):
        import numpy as np
        import ia636 as ia
        vmin,vmax = np.percentile(f,[p, 100-p])
        g = np.clip(f,vmin,vmax)
        gn = np.rint(255. * (g - vmin) / (vmax-vmin))
        #return ia.ianormalize(g)    
        return gn
    
    # -----------------------------------
    # Registra cada procedimento de teste
    #     def addTester(descr, fun, args=[])
    # -----------------------------------
    
    a = np.array([0,2,3,3,4,4,4,5,5,7])
    print nn_gabarito(a,perc=10)
    f2 = adreadgray('MVBook/cells_yeast.jpg')
    tester.add_test_case([a,10], nn_gabarito(a,10), label='numerico', show_in_results_table=True)
    tester.add_test_case([arange(256),10],  nn_gabarito(arange(256),10) , label='rampa', show_in_results_table=True)
    tester.add_test_case([f2,10],  nn_gabarito(f2,10) , label='imagem', show_in_results_table=True)
    
    tester.add_train_case([a,10], nn_gabarito(a,10))
    tester.add_train_case([arange(256),10],  nn_gabarito(arange(256),10))
    tester.add_train_case([f2,10], nn_gabarito(f2,10) )
