Criando o teste

.. code:: python module
    
    import numpy as np
    from tbx_perftest.tester import Tester
    
    def assert_fnc(user_fnc, input, groundtruth):
        X,n = input
        Xnew = groundtruth
        xnew = user_fnc(X,n)
        err = np.finfo(np.float32).eps
        c = abs(xnew/Xnew)
        ok = ((c > 1-err) & (c < 1+err)).all()
        return ok
        #return (np.abs(  Xnew  - xnew).max() < 1.) or \
        #       (np.abs((-Xnew) - xnew).max() < 1.)    

    tester = Tester(assert_fnc, 'tbx_perftest', 'pirp_4e_pca', runs=3)

    from tbx_perftest.pirp_4e_pca import tester
    import numpy as np
    import ia636 as ia
    
    # IMPORTANTE: Os casos de teste devem ser apagados se esta página sofrer múltiplas edições
    tester.reset_cases()
    
    def pca_gabarito(X,n):
        V,sigma,mu = ia.iapca(X)
        return np.dot(X-mu,V[:,:n])
    
    # -----------------------------------
    # Registra cada procedimento de teste
    #     def addTester(descr, fun, args=[])
    # -----------------------------------
    
    X0 = array([[0,0,0],
                [1,2,3],
                [3,5,8],
                [4,7,15]])
    yX_train = adreadgray('ia636/digits_train.png')
    casos = [\
         [(X0,2),            '(4,3) n=2'],
         [(yX_train[:,1:],2),'digitos 2-D'],
         ]
    
    for a,msg in casos:
        tester.add_test_case(a, pca_gabarito(*a), label=msg, show_in_results_table=True)
        tester.add_train_case(a, pca_gabarito(*a))
