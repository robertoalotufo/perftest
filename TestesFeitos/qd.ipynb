    from tbx_perftest.tester import Tester
    import numpy as np
    
    def assert_fnc(user_fnc, input, groundtruth):
        f,step = input
        f_save = f.copy()
        OK      = float(np.array_equal(user_fnc(f,step),groundtruth))
        OK_copy = float(np.array_equal(f,f_save))
        return (OK+OK_copy)/2
    
    tester = Tester(assert_fnc, 'tbx_perftest', 'pirp_4e_qd',3)
